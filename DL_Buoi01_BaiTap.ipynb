```python
#12
@tf.function
def calculator(a, b):
    return a + b, a - b, a * b, a / b

x12 = tf.random.uniform([3, 2, 2])
y12 = tf.random.uniform([3, 2, 2])
res_add, res_sub, res_mul, res_div = calculator(x12, y12)
print(res_add.numpy(), res_sub.numpy(), res_mul.numpy(), res_div.numpy(), "\n")
```

```python
#13
class TotalModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return x + y

model13 = TotalModel()
print(model13(tf.constant(5), tf.constant(3)).numpy(), "\n")
```

```python
#14
class SubMulModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return x - y, x * y

model14 = SubMulModel()
res_sub, res_mul = model14(tf.constant(10), tf.constant(2))
print(res_sub.numpy(), res_mul.numpy(), "\n")
```

```python
#15
class PerimeterModel(tf.Module):
    @tf.function
    def __call__(self, a, b):
        return (a + b) * 2

model15 = PerimeterModel()
print(model15(tf.constant(4.0), tf.constant(5.0)).numpy(), "\n")
```

```python
#16
class SurfaceAreaModel(tf.Module):
    @tf.function
    def __call__(self, a, b, c):
        s_xq = (a + b) * c * 2
        s_tp = (a * b) * 2 + s_xq
        return s_tp

model16 = SurfaceAreaModel()
print(model16(tf.constant(2.0), tf.constant(3.0), tf.constant(4.0)).numpy(), "\n")
```

```python
#17
class Proof17(tf.Module):
    @tf.function
    def __call__(self, ad, bc, ab, cd):
        condition = tf.logical_and(tf.equal(ad, bc), tf.equal(ab, cd))
        return tf.where(condition, "AD // BC", "Không song song")

model17 = Proof17()
print(model17(tf.constant(5), tf.constant(5), tf.constant(3), tf.constant(3)).numpy().decode(), "\n")
```

```python
#18
class Proof18(tf.Module):
    @tf.function
    def __call__(self, ad_len, bc_len, is_parallel):
        condition = tf.logical_and(tf.equal(ad_len, bc_len), is_parallel)
        return tf.where(condition, "AB = CD", "Không bằng nhau")

model18 = Proof18()
print(model18(tf.constant(10), tf.constant(10), tf.constant(True)).numpy().decode(), "\n")
```

```python
#19
class Proof19(tf.Module):
    @tf.function
    def __call__(self, mn_len, mb_len, is_parallel):
        condition = tf.logical_and(tf.equal(mn_len, mb_len), is_parallel)
        return tf.where(condition, "NA = NC", "Không bằng nhau")

model19 = Proof19()
print(model19(tf.constant(5), tf.constant(5), tf.constant(True)).numpy().decode())
```
